```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
file = "Resources/union_settlements_1995.csv"
data_csv = pd.read_csv(file)
```

```python
df = pd.DataFrame(data_csv)
df.info()
df.sample(n=5)
```

```python
# List columns
df.columns
```

```python
# Get total settlements by union
union_count = df.UNION.count()
print(union_count)
```

```python
# df = df.dropna(how='any')
# df.plot()
plot_filter = df[["NUMBER OF EMPLOYEES", "UNION"]]
plot_filter.plot()
plt.show()
```

```python
df.plot()
plt.show()
```

```python
# Configure plot, figsize, title, and axis labels
plt.title("Unions R Bad Mkay")
plt.xlabel("Settlements")
plt.ylabel(df.UNION)
plt.bar(union_count, df.UNION)
plt.show()

# Configure x-tick rotation


# Show plot
```
