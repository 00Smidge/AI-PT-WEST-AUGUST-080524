```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
csv_data = pd.read_csv("../Resources/avg_rain_state.csv")
rain_df = pd.DataFrame(csv_data)
```

```python
rain_df.info()
rain_df.sample(n=5)
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(rain_df))
tic_loc = [value + 0.4 for value in x_axis]
plt.xticks(x_axis, tic_loc)
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
states = rain_df.State
inches = rain_df.Inches
plt.figure(figsize=(10, 8))
plt.bar(states, inches, color="#eda569", alpha=0.5, align="edge")
plt.xticks(rotation="vertical")
plt.show()
```
