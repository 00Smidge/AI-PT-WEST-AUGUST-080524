```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis to numerical value for month
x_axis = np.arange(1,13,1)
```

```python
# Avearge weather temp
points_F = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Convert to Celsius C = (F-32) * 0.56
points_C = []
for temp in points_F:
  points_C.append((temp-32)*.56)
```

```python
# Create a handle for each plot
line_1 = plt.plot(x_axis, points_F, marker='X', color="#127f9f", label="Temp (F)")
line_2 = plt.plot(x_axis, points_C, marker='s', color="#eda569", label="Temp (C)")
plt.legend()
plt.show()
```
