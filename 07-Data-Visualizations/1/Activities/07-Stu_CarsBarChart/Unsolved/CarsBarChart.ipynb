```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["San Francisco", "Omaha", "New Orleans", "Cincinnati", "Pittsburgh"]
cars_in_cities = [214.7, 564.4, 416.5, 466.7, 350.6]
x_axis = np.arange(len(cars_in_cities))
```

```python
# Create a bar chart based upon the above data
plt.bar(cities, cars_in_cities, color='#eda569', align='center', alpha=0.5)
```

```python
# Create the ticks for our bar chart's x axis
x_ticks = plt.xticks(x_axis)
```

```python
# Set the limits of the x axis
x_tick_lims = plt.xlim(-1, len(x_axis))
```

```python
# Set the limits of the y axis
y_tick_lims = plt.ylim(0, max(cars_in_cities)+75)
# plt.show()
```

```python
# Give the chart a title, x label, and y label
plt.title('Personal Vehicles per capita per city')
plt.xlabel('Cities')
plt.ylabel('Autos Per Capita')
```

```python
# Save an image of the chart and print it to the screen
plt.show()
plt.savefig('first_fig.webp')
```
