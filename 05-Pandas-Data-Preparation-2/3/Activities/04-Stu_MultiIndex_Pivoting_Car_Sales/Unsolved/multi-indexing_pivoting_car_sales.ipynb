```python
# Import dependencies
import pandas as pd
data = 'Resources/vehicles.csv'
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv(data)
df.info()
```

```python
df = df.rename(columns={'make':'model', 'model':'make'})
```

```python
# Show the total number of vehicles for each model and make.
# Show the top 20 results.
df[['make','model']].value_counts().head(20)
```

```python
# Rename the "count" column "Total"
df = df.rename(columns={'count':'Total'})
```

```python
# Sort the renamed pivot table on the "Total" column to show the top 20 vehicles sold by model and make. 
top_autos = pd.pivot_table(df, index=['make','model'], values='Total', aggfunc='sum')
top_autos.sort_values(by='Total', ascending=False).head(20)
```

```python
# Show the total number of vehicles for each model, make, and year.
total_each = pd.pivot_table(df, index=['make','model','year'], values='Total', aggfunc='sum')
# Show the top 20 results.
total_each.head(20)
```

```python
# Sort the pivot table on the "year" index and "Total" column and show the top 20 vehicles sold by the year.
total_each.sort_values(by=['year', 'Total'], ascending=False).head(20)
```

```python
# Show the total number of vehicles for each category and year. 
year_of_cats = pd.pivot_table(df,
                              index=['category', 'year'],
                              values=('Total'),
                              aggfunc='sum')
# Show the top 20 results.
year_of_cats.head(20)
```

```python
# Sort the pivot table on the on the "Total" column to show the top 20 vehicles sold by category and year.
sort_ttls = year_of_cats.sort_values(by='Total', ascending=False)
sort_ttls.head(20)
```

```python
# Show the total number vehicles for each category, model, and year.
ttl_autos_yrs = df.pivot_table(index=['category', 'make', 'year'], values='Total', aggfunc='sum')

# Show the top 20 results.
ttl_autos_yrs.head(20)
```

```python
# Sort the pivot table on the on the "Total" column to show the top 20 vehicles sold by category, model, and year.
sort_yr_ttls = ttl_autos_yrs.sort_values(by='Total', ascending=False)
sort_yr_ttls.head(20)
```
