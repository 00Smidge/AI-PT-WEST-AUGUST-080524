```python
# Import the required libraries and dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the data from the tsla_historical.csv file into a Pandas DataFrame
df_tsla = pd.read_csv("../Resources/tsla_historical.csv")
```

```python
# Display the first five rows of the DataFrame
df_tsla.head()
```

```python
# Inspect the DataFrame's data types using the info function
df_tsla.info()
```

```python
# Transform the time column to a datetime data type
df_tsla["time"] = pd.to_datetime(df_tsla["time"], utc=True)
```

```python
# Display the first five rows of the DataFrame to confirm
# changes to the time column
df_tsla.head()
```

```python
# Use the info function to confirm the change in data type
# for the time column
df_tsla.info()
```

```python
# Set the time column as DataFrame index
df_tsla = df_tsla.reset_index(col_level="time")  # YOUR CODE HERE

# Display the first five rows of the DataFrame
# YOUR CODE HERE
```

```python
# Plot the closing price of TSLA
# YOUR CODE HERE
plt.figure(figsize=(18, 12))
plt.plot(df_tsla.time, df_tsla.close)
plt.xlabel("time")
plt.ylabel("Closed at")
plt.title("Historical TSLA Closing Prices")
plt.tight_layout()
plt.show()
```

```python
# Select and plot the TSLA closing prices from 2020
# YOUR CODE HERE
from_2020 = df_tsla[df_tsla["time"] > "2020"]
```

```python
plt.figure(figsize=(18, 12))
plt.plot(from_2020.time, from_2020.close)
plt.xlabel("time")
plt.ylabel("Closed at")
plt.title("Historical TSLA Closing Prices")
plt.tight_layout()
plt.show()
```

```python
# Select and plot the TSLA closing prices from 2020 using DatetimeIndex attributes
# YOUR CODE HERE
plt.figure(figsize=(18, 12))
plt.plot(from_2020.time.index, from_2020.close)
plt.xlabel("time")
plt.ylabel("Closed at")
plt.title("Historical TSLA Closing Prices")
plt.tight_layout()
plt.show()
```
