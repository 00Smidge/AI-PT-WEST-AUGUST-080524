```python
# Import necessary libraries and dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import data
sp500_path = "../Resources/sp500_stock_volume.csv"

# Read the S&P 500 volume into a DataFrame. (Make sure to declare the datetime index).
sp_csv = pd.read_csv(sp500_path, index_col="Date").reset_index()
sp_df = pd.DataFrame(sp_csv)
sp_df.Date = pd.to_datetime(sp_df.Date)
sp_df["weekday"] = sp_df.Date.dt.strftime("%A")
```

```python
# Display sample data
sp_df.sample(n=5)
```

```python
# Slice the dataframe so that it just includes the volume data.
sp_df_vols = sp_df["volume"]
```

```python
# Display sample data
sp_df_vols.sample(n=5)
```

```python
# Declare the group level to be the day of the week (e.g., Mon, Tues,...)
grouped = sp_df.groupby('weekday')['volume'].sum()
grouped = grouped.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
```

```python
grouped
```

```python


# Plot average daily volume according to day of the week
plt.figure(figsize=(18, 12))
plt.bar(grouped.index, grouped.values, color='#eda569')
plt.xlabel("Day of the Week")
plt.ylabel("Average Daily Volume")
plt.title("Average Daily Trading Volume by Day of the Week")
plt.show()
```
